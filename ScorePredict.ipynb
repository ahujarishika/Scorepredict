```
#STEP 1: IMPORT PYTHON LIBRARIES REQUIRED TO PERFORM PREDICTION
import pandas as pd
import numpy as np  
from sklearn.model_selection import train_test_split  
from sklearn.linear_model import LinearRegression  
from sklearn import metrics  
import matplotlib.pyplot as plt  
#MAKING SURE THE GRAPHS PLOTTED ARE ININE WITH THE CODE
%matplotlib inline
```

```
#STEP 2: IMPORT DATA FROM LINK AND DISPLAY
url = "http://bit.ly/w-data"
#ASKING PANDAS TO READ THE CSV(COMMA SEPARATED VALUES) FILE FROM THE URL
s_data = pd.read_csv(url)
print("Data imported successfully")
#DISPLAYING THE DATA AND ITS DETAILS
print(s_data)
s_data.info()
```

```
#STEP 3:PLOTTING SCORES VS HOURS IN A GRAPH TO SHOW LINEARITY 
#PLOT THE GRAPH WITH HOURS ON THE X AXIS AND SCORES ON THE Y AXIS
s_data.plot(x='Hours', y='Scores', style='*')  
#GIVE THE PLOT TITLE AND NAME THE X AND Y AXIS'
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Scored')  
#DISPLAY GRAPH
plt.show()
```

```
#STEP 4:SEPARATE HOURS AND SCORES INTO DIFFERENT ARRAYS
#EXTRACT DATA FROM ALL ROWS ALL COLUMNS EXCEPT THE LAST ONE TO GET HOURS
X = s_data.iloc[:, :-1].values  
#EXTRACT DATA FROM ALL ROWS AND THE FIRST COLUMN TO GET PERCENTAGE
y = s_data.iloc[:, 1].values  
```

```
#STEP 5:SPLIT THE DATA FOR TRAINING AND TESTING USING Scikit-Learn's BUILT-IN train_test_split() METHOD  
#20% DATA IS RESERVED FOR TESTING AND 80% IS TAKEN FOR TRAINING
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=0)
```

```
#STEP 6:TRAINING THE DATA USING LINEAR REGRESSION
#OBJECT FOR THE FUNCTION
regressor = LinearRegression()  
#PASS THE DATA RESERVED FOR TRAINING TO THE FUNCTION
regressor.fit(X_train, y_train) 

print("Training complete.")
```

```
#STEP 7:PLOTTING THE REGRESSION LINE TO SHOW ALL POINTS ARE AT MINIMUM DISTANCE FROM IT
#y= mX + C
line = regressor.coef_*X+regressor.intercept_

# PLOTTING FOR THE TEST DATA
plt.scatter(X, y)
plt.plot(X, line);
plt.show()
```

```
#STEP 8:ATTAIN PREDICTED VALUES FOR TESTING DATA
# Testing data - In Hours
print("Testing values:",X_test) 
# Predicting the scores
y_pred = regressor.predict(X_test)
print("Predicted perecentages",y_pred)
```

```
#STEP 9:COMPARING ACTUAL VS PREDICTED
#PRINT THE ACTUAL DATA VS THE PREDICTED DATA
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df 
```

```
#STEP 10: PREDICT THE PERCENTAGE FOR A STUDENT THAT STUDIES FOR 9.25 HOURS
hours = [9.25]
own_pred = regressor.predict([hours])
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```

```
#STEP 11: EVALUATING THE ALGORITHM BY CALCULATING ERROR
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test, y_pred)) 
```

**Hence a student that studies for 9.25 hours is predicted to score 93.69173248737539**
